```
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO
```

```
"""## 2. Data Loading and Initial Exploration"""
```

```
# Load the dataset
df = pd.read_csv('Crop_recommendation.csv')
```

```
print("First 5 rows of the DataFrame:")
print(df.head())
```

```
print("\nDataFrame Info:")
df.info()
```

```
print("\nDescriptive Statistics:")
print(df.describe())
```

```
"""## 3. Data Preprocessing"""
```

```
# Separate features (X) and target (y)
X = df.drop('label', axis=1)
y = df['label']
```

```
# Encode the target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```
"""## 4. Model Training and Evaluation"""
```

```
# Initialize and train the XGBoost Classifier
model = xgb.XGBClassifier(objective='multi:softmax', num_class=len(le.classes_), random_state=42)
model.fit(X_train, y_train)
```

```
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=le.classes_)
```

```
print(f"Accuracy Score: {accuracy:.4f}")
print("\nClassification Report:\n", report)
```

```
"""### Confusion Matrix Visualization"""
```

```
# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```
# Plot the confusion matrix
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```
"""## 5. Prediction Function"""
```

```
def predict_crop(N, P, K, temperature, humidity, ph, rainfall):
    # Create a DataFrame from the input values, ensuring the correct order of features
    input_data = pd.DataFrame([[N, P, K, temperature, humidity, ph, rainfall]],
                              columns=X.columns) # Use X.columns to maintain feature order

    # Make prediction
    prediction_encoded = model.predict(input_data)

    # Inverse transform the prediction to get the crop name
    predicted_crop = le.inverse_transform(prediction_encoded)

    return predicted_crop[0]
```

The classification_report is already available as 'report' in string format.
We need to parse this string into a DataFrame for better display.

```
report_data = StringIO(report)
```

Read the string into a DataFrame, skipping initial lines if needed and handling potential extra columns.
The classification report often has an empty first column for the index labels, and 'support' at the end.
We will extract the relevant metrics for each class.

```
# Read the report line by line to process it
lines = report.split('\n')
```

```
# Find the header line for metrics (precision, recall, f1-score, support)
header_index = -1
for i, line in enumerate(lines):
    if 'precision' in line and 'recall' in line:
        header_index = i
        break
```

```
if header_index != -1:
    # Extract the header line and data lines
    header_line = lines[header_index].strip().split()
    data_lines = [line.strip().split() for line in lines[header_index+1:] if line.strip() and not line.strip().startswith(('accuracy', 'macro avg', 'weighted avg'))]

    # Create a DataFrame
    if data_lines:
        # Ensure all data rows have the same number of columns as the header, plus one for the label
        processed_data = []
        for row in data_lines:
            # The first element is the class label, the rest are metrics
            label = row[0]
            metrics = [float(x) for x in row[1:]]
            processed_data.append([label] + metrics)

        # Create columns: 'class' and then the header_line items
        columns = ['class'] + header_line
        df_report = pd.DataFrame(processed_data, columns=columns)
        display(df_report)
    else:
        print("No class-specific data found in the classification report.")
else:
    print("Could not parse the classification report.")
```
